## Prolem with perceptron


*   The problem with percepteron is that its don't work on non-linear data its only work on linear data which show in this example.



```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
and_data=pd.DataFrame({'input1':[0,0,1,1],'input2':[0,1,0,1],'output':[0,0,0,1]})
or_data=pd.DataFrame({'input1':[0,0,1,1],'input2':[0,1,0,1],'output':[0,1,1,1]})
xor_data=pd.DataFrame({'input1':[0,0,1,1],'input2':[0,1,0,1],'output':[0,1,1,0]})
```

```python
and_data
```

```python
or_data
```

```python
xor_data
```

```python
# and data output
sns.scatterplot(x='input1',y='input2',data=and_data,hue='output',s=200)
```

```python
# or data output
sns.scatterplot(x='input1',y='input2',data=or_data,hue='output',s=200)
```

```python
# xor data output
sns.scatterplot(x='input1',y='input2',data=xor_data,hue='output',s=200)
```

```python
from sklearn.linear_model import Perceptron
p1=Perceptron()
p2=Perceptron()
p3=Perceptron()

p1.fit(and_data[['input1','input2']],and_data['output'])
p2.fit(or_data[['input1','input2']],or_data['output'])
p3.fit(xor_data[['input1','input2']],xor_data['output'])
```

```python
p1.coef_ , p1.intercept_
```

```python
p2.coef_ , p2.intercept_
```

```python
p3.coef_ , p3.intercept_
```
