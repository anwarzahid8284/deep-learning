## Perceptron Trick and Training

**Multiple Linear Regressiom Model**
*   y=w0X0+w1X1+w2X2

```python
# make a classification data set
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define dataset
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_informative=1,n_redundant=0, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
# green represent class 1 and blue represent class 0
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.show()
```

```python
# percepteron algorithm for to find w0,w1,w2
# mathematical model
# z= w_not*x_not+w_one*x1,w_two*x2

def perceptronn(X,y):
    X=np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1],weights[2]
```

```python
# step function
def step(z):
    return 1 if z>0 else 0
```

```python
# wights and bais
intercept_,w1,w2 = perceptronn(X,y)
```

```python
# Ax+By+C=0
# m = -A/B ,  c= -C/B where A=w1, B=w2 and C=intercept
m = -w1/w2
c = -intercept_/w2
```

```python
# draw a line line
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.plot(x_input,y_input,color='red')
plt.ylim(-3,2)
```
