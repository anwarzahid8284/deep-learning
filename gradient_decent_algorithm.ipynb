**Gradient Descent Using Linear Regression**

Gradient descent is an iterative optimization algorithm used in linear regression to find the best-fitting line by minimizing the cost function.

**Slope Formula**

Determines the steepness and direction of the line.
A larger magnitude means the output changes more for small input changes.
$$
\text{slope} = -2 \sum_{i=1}^{n} (y_i - m x_i - b)
$$

**Bias Formula**

Shifts the line up or down on the y-axis.
It adjusts the output when input x=0.

$$
b_{\text{new}} = b_{\text{old}} - \text{learning_rate} \times \text{slope}
$$

**Learning Rate**

 learning rate (controls how big each step)

**Line Representation**

1. Red line represent best fit line (which is our requirement to get using gradienct decent)
2. Green line represent recent line
3. Gray line represent all previous line

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
# plot linear data in 2D graph
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

**Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X,y)
```

```python
# slop m value after train a model
reg.coef_
```

```python
# intercept b vlaue for which model gives us best fit line means the line having give a model minimum error
reg.intercept_
```

**Best Fit Line**

```python
# this is the best fit line which give us minimum error
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

**Apply Gradienct Decent**

1. Lets apply Gradient Descent assuming slope is constant m = 78.35
2. and let's assume the starting value for intercept b = 0

```python
# first find the value of y to plot line
# y=mx+b to get value for yi
y_pred = ((78.35 * X) + 0).reshape(4)
```

**Initial Line**

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
```

**First Itration**

```python
# calculate slope
m=78.35
b=0
slope = -2 * np.sum(y-m*X.ravel()-b)
slope
```

```python
# let learning rate 0.1 and calcualte step size
lr=0.1
step_size=lr*slope
step_size
```

```python
# calculate new bias
b=b-step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#D3D3D3',label='b = 0')
plt.legend()
plt.show()
```

**Gradient Decent Using Loop Approach**

The above step is just for understanding what behind its work but now i use loop aproach

```python
b_loop = 0
m_loop = 78.35
lr_loop = 0.01

epochs = 5

for i in range(epochs):
  sloop_loop= -2 * np.sum(y - m_loop*X.ravel() - b_loop)
  b = b - (lr_loop * sloop_loop)

  y_pred_loop = m_loop * X + b_loop

  plt.plot(X,y_pred_loop)

plt.scatter(X,y)
```
